```python
x = list(range(-10,10))
y = [(n ** 2) - n + 2 for n in x]

plt.scatter(x,y)
plt.annotate('origin', xy=(0, 0), xytext=(0, 20),
             arrowprops={'facecolor': 'black'})
plt.plot(x,y,ls=':')
plt.scatter(x, y)

plt.title('Question 1: $y = {x}^2 - x + 2$')
plt.xlabel('$x$')
plt.ylabel('${x}^2 - x + 2$')

plt.show()
```

```python
x = list(range(10))
y = [math.tan(n) for n in x]

plt.plot(x,y,ls='--')
plt.scatter(x,y)

plt.title('Question 2c: $y = \ tan (x)$')
plt.xlabel('$x$')
plt.ylabel('$\ tan (x)$')

plt.show()
```

4. Combine the figures you created in the last step into one large figure with 4 subplots

```python
plt.figure(figsize=(14, 10))
plt.suptitle('4 separate plots for the listed equations')

x = list(range(5))
y1 = [math.sqrt(n) for n in x]
y2 = [math.pow(3, n) for n in x]

y3 = [math.tan(n) for n in x]
y4 = [math.pow(n, 2) for n in x]

plt.subplot(221)
plt.scatter(x, y1,marker='X')
plt.plot(x, y1,ls='--')
plt.title('Question 2a: $y = \sqrt{x}$')

plt.subplot(222)
plt.scatter(x, y2,marker='X')
plt.plot(x, y2,ls='--')
plt.title('Question 2b: $y = {x}^3$')

plt.subplot(223)
plt.scatter(x, y3,marker='X')
plt.plot(x, y3,ls='--')
plt.title('Question 2c: $y = \ tan (x)$')

plt.subplot(224)
plt.scatter(x, y4,marker='X')
plt.plot(x, y4,ls='--')
plt.title('Question 2d: $y = {2}^x$')

plt.show()
```

5. Comine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend.

```python
x = list(range(5))
y1 = [math.sqrt(n) for n in x]
y2 = [math.pow(3, n) for n in x]
y3 = [math.tan(n) for n in x]
y4 = [math.pow(n, 2) for n in x]

plt.plot(x, y1, c='red', label = '$y = \sqrt{x}$')
plt.plot(x, y2, c='blue', label = '$y = {x}^3$')
plt.plot(x, y3, c='orange', label = '$y = \ tan (x)$')
plt.plot(x, y4, c='black', label = '$y = {2}^x$')

plt.legend(loc='best')
plt.show()
```
